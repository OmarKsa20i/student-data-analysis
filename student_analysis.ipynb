```python
data['Average'] = data[['G1', 'G2', 'G3']].mean(axis=1)
top_student = data.loc[data["Average"].idxmax()]
print("\n Top Student:")
print(top_student)
```

```python
import matplotlib.pyplot as plt


top_10_students = data.sort_values(by='Average', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(top_10_students.index.astype(str), top_10_students["Average"], color='skyblue')
plt.title("Average Scores of Top 10 Students")
plt.xlabel("Student Index")
plt.ylabel("Average Score")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
